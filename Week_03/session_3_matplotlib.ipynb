### 23. Populate Subplots (Part 1):

### 26. Plotting an ML Function: Plot the Sigmoid activation function.
**Formula:** `y = 1 / (1 + e^(-x))`
- Generate `x = np.linspace(-10, 10, 100)`.
- Generate `y` using the formula.
- Plot `y` vs. `x`. Add a title "Sigmoid Function".
- Add a horizontal grid line to show where the function crosses 0.5.

**Hint:** Use `np.exp()` for the exponential. For the grid line, use `ax.grid(axis='y', linestyle='--')`.

```python
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))

fig, ax = plt.subplots(figsize = (6,4))
ax.set_title("Sigmoid Function")
ax.plot(x,y)
ax.axhline(y=0.5, color='r', linestyle='--', label='y=0.5 threshold')
ax.grid(axis = 'x', linestyle = "--")
plt.legend()
plt.show()
```

### 28. Visualizing Model Performance (Loss Curve):
**Scenario:** A common task in ML is to plot the training and validation loss over epochs to see if the model is learning well or overfitting.
- **Data:**
  `epochs = np.arange(1, 21)`
  `train_loss = 1 / epochs + np.random.rand(20) * 0.1`
  `val_loss = 1.2 / epochs + np.random.rand(20) * 0.1 + 0.05`
- Plot `train_loss` vs. `epochs` (label='Training Loss', color='blue').
- On the same axes, plot `val_loss` vs. `epochs` (label='Validation Loss', color='orange').
- Set the x-label to `Epochs`, y-label to `Loss`, and add a legend.

```python
epochs = np.arange(1,21)
train_loss = 1/ epochs + np.random.rand(20) * 0.1
val_loss = 1.2 / epochs + np.random.rand(20) * 0.1 + 0.05

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(epochs,train_loss,label = "Training Loss",color = "blue")
ax.plot(epochs,val_loss,label = "Validating Loss",color = "orange")
ax.set_xlabel("Epochs")
ax.set_ylabel("Loss")
plt.legend()
plt.show()
```

### 29. Visualizing ML Clusters:
**Scenario:** Visualizing how a clustering algorithm (like K-Means) has grouped data points.
- **Data: Create two "blobs" of data:**
  `blob1_x = np.random.normal(2, 1, 100)`
  `blob1_y = np.random.normal(5, 1, 100)`
  `blob2_x = np.random.normal(6, 1, 100)`
  `blob2_y = np.random.normal(2, 1, 100)`
- Plot `blob1` as a scatter plot (color='purple', label='Cluster 1').
- On the same axes, plot `blob2` as a scatter plot (color='red', label='Cluster 2').
- Add a legend.

```python
blob1_x = np.random.normal(2,1,100)
blob1_y = np.random.normal(5,1,100)
blob2_x = np.random.normal(6,1,100)
blob2_y = np.random.normal(2,1,100)

fig, ax = plt.subplots(figsize = (6,4))
ax.scatter(blob1_x,blob1_y,color = "purple" , label = "Cluster 1" , alpha=0.5 , s = 50)
ax.scatter(blob2_x,blob2_y,color = "red" , label = "Cluster 2" , alpha=0.5 , s = 50)
plt.legend()
plt.show()
```
